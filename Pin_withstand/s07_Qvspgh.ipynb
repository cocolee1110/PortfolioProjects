```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
def get_avg_flow_and_samples(filename):
    df = pd.read_csv(filename, skiprows=14, usecols=['Relative Time[s]', 'Flow [ul/min]']) 
    avg_flow = df["Flow [ul/min]"].mean() - 0.01683716564992958
    num_samples_count = df.shape[0]
    return avg_flow, num_samples_count
```

```python
h_s07_2x2 = [2,5,7,10,15]
trials = [1,2,3]
pgh_s07_2x2 = [1000 * 9.81 * i * 1E-2 * 1/133 for i in h_s07_2x2] * 3
avg_flows_s07_2x2 = []
num_samples = []

for n in trials:
    for i in h_s07_2x2:
        filename = f"s07_2x2/s07_2x2_{i}cm_trial{n}_withstandPa.csv"
        avg_flow_s07_2x2, num_samples_count = get_avg_flow_and_samples(filename)
        avg_flows_s07_2x2.append(avg_flow_s07_2x2)
        num_samples.append(num_samples_count)

avg_flows_s07_2x2

```

```python

h_s07_3x2 = [2,5,7,10,15]
trials = [1,2,3]
pgh_s07_3x2 = [1000 * 9.81 * i* 1E-2 * 1/133 for i in h_s07_3x2] *3
avg_flows_s07_3x2 = []
num_samples = []

for n in trials:
    for i in h_s07_3x2:
        filename = f"s07_3x2/s07_3x2_{i}cm_trial{n}_withstandPa.csv"
        avg_flow_s07_3x2, num_samples_count = get_avg_flow_and_samples(filename)
        avg_flows_s07_3x2.append(avg_flow_s07_3x2)
        num_samples.append(num_samples_count)


avg_flows_s07_3x2
```

```python

h_s07_3x15 = [2,3,5,7,10,15]
trials = [1,2,3]
pgh_s07_3x15 = [1000 * 9.81 * i * 1E-2 *1/133 for i in h_s07_3x15] *3
avg_flows_s07_3x15 = []
num_samples = []

for n in trials:
    for i in h_s07_3x15:
        filename = f"s07_3x1.5/s07_3x15_{i}cm_trial{n}_withstandPa.csv"
        avg_flow_s07_3x15, num_samples_count = get_avg_flow_and_samples(filename)
        avg_flows_s07_3x15.append(avg_flow_s07_3x15)
        num_samples.append(num_samples_count)


avg_flows_s07_3x15
```

```python

h_s07_2x15 = [2,5,7,10,15]
trials = [1,2,3]
pgh_s07_2x15 = [1000 * 9.81 * i * 1E-2 * 1/133 for i in h_s07_2x15] * 3
avg_flows_s07_2x15 = []
num_samples = []

for n in trials:
    for i in h_s07_2x15:
        filename = f"s07_2x15/s07_2x15_{i}cm_trial{n}_withstandPa.csv"
        avg_flow_s07_2x15, num_samples_count = get_avg_flow_and_samples(filename)
        avg_flows_s07_2x15.append(avg_flow_s07_2x15)
        num_samples.append(num_samples_count)

avg_flows_s07_2x15
```

```python
x1 = pgh_s07_2x2
y1 = avg_flows_s07_2x2

x2 = pgh_s07_3x2
y2 = avg_flows_s07_3x2

x3 = pgh_s07_3x15
y3 = avg_flows_s07_3x15

x4 = pgh_s07_2x15
y4 = avg_flows_s07_2x15

x1 = np.array(x1)
y1 = np.array(y1)
x2 = np.array(x2)
y2 = np.array(y2)
x3 = np.array(x3)
y3 = np.array(y3)
x4 = np.array(x4) 
y4 = np.array(y4) 

# Fit the data to a line (y = mx + c)
m1, c1 = np.polyfit(x1, y1, 1)
m2, c2 = np.polyfit(x2, y2, 1)
m3, c3 = np.polyfit(x3, y3, 1)
m4, c4 = np.polyfit(x4, y4, 1) 

# Compute the values of the line at the positions of the data
y1_fit = m1*x1 + c1
y2_fit = m2*x2 + c2
y3_fit = m3*x3 + c3
y4_fit = m4*x4 + c4  # y-values of fit for new data

# Plot the data and the fit
plt.figure(figsize=(20, 10))
plt.plot(x1, y1, 'o', color='r', label='2x2')
plt.plot(x2, y2, 'o', color='g', label='3x2')
plt.plot(x3, y3, 'o', color='b', label='3x1.5')
plt.plot(x4, y4, 'o', color='purple', label='2x1.5') 
plt.plot(x1, y1_fit, '-', color='r')
plt.plot(x2, y2_fit, '-', color='g')
plt.plot(x3, y3_fit, '-', color='b')
plt.plot(x4, y4_fit, '-', color='purple') 


plt.grid(True)
plt.title('s07 Q vs pgh')
plt.xlabel('Pressure difference (mmHg)')
plt.ylabel('Q (ul/min)')
plt.legend(loc= 'best')
plt.show()
```
